```python
# Import numpy and pandas packages
import numpy as np
import pandas as pd
```

```python
# Read the data set cars
cars = pd.read_csv("cars.csv")
```

```python
cars
```

```python
# find correlation for all variables in cars
cars.corr()
```

```python
# (cyl and mpg) and (wt and mpg) have high negative correlations. (cyl and disp) and (wt and disp) have high positive correlations
# create a scatter plot to see relationship between hp and mpg
cars.plot.scatter(x = "hp", y = "mpg")
```

```python
# there seems to be a negative correlation between hp and mpg.
# import a new package to use linear models
from sklearn import linear_model
```

```python
# create a 2d array to allow linear regression to run
X = np.array(cars['hp']).reshape(-1,1)
Y = np.array(cars['mpg']).reshape(-1,1)
```

```python
# create value regr to simplify the code
regr = linear_model.LinearRegression()
regr.fit(X,Y)
```

```python
# represents the coefficient of our hp variable
regr.coef_
```

```python
# represents the y-intercept of our linear regression model
regr.intercept_
```

```python
# represents our R-squared value for our model. In this case 60 percent of mpg can be explained by our model.
regr.score(X,Y)
```

```python
# linear regression formula. mpg = 30.0988 - 0.06822hp + e
# predict what value of mpg the car has if it has 100 hp
regr.predict([[100]])
```

```python
# Our linear regression model predicted a car that gets 23.276 mpg based on a horsepower of 100. 
# import subpackage to allow mean absolute error calculation
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y, regr.predict(X))
```

```python
# The mean absolute error represents the average of the absolute value of the differences between the measured 
# observations and actual observations. This means that on average our model has an error of plus or minus 2.9074 mpg. 
```
